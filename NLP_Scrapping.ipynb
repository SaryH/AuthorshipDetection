```python
# import necessery libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Scrap poems name from www.aldiwan.net website

"""
# Links to scrap poems from
https://www.aldiwan.net/cat-poet-farazdaq
https://www.aldiwan.net/cat-poet-albohtry
https://www.aldiwan.net/cat-poet-safi-al-din-al-hilli
https://www.aldiwan.net/cat-poet-Ibn-Arabi
https://www.aldiwan.net/cat-poet-ahmed-shawqi
"""

url = "https://www.aldiwan.net/cat-poet-ahmed-shawqi"  # Replace with the URL of the web page you want to scrape
response = requests.get(url)
content = response.content

soup = BeautifulSoup(content, "html.parser")

parent_elements = soup.find_all("div", class_="content row bb bb-none-1")

names = []
poem_name = []

# Extract poems name from the website
for parent_element in parent_elements:
    names.extend(parent_element.find_all("a", class_="float-right"))

for name in names:
    poem_name.append(name.text)
```

```python
# display first 10 poems name
poem_name[:10]
```

```python
# Scrap document names

url = "https://www.aldiwan.net/cat-poet-ahmed-shawqi"  # Replace with the URL of the web page you want to scrape
response = requests.get(url)
content = response.content

soup = BeautifulSoup(content, "html.parser")
links = soup.find_all("a")

href_links = [link.get("href") for link in links]

links = []
counter = 0
# Process and use the extracted links as needed
for link in href_links:
    if str(link)[0:4] == 'poem':
        links.append(link)
        counter += 1
```

```python
# remove duplicated documents names with order saving
def unique(seq):
    seen = set()
    seen_add = seen.add
    return [x for x in seq if not (x in seen or seen_add(x))]

links = unique(links)
```

```python
# display first 10 document names
links[:10]
```

```python
# Scrap poems content

poems = []
timeout_seconds = 15

for i in range(len(links)):
    # URL of the website containing the poem
    url = "https://www.aldiwan.net/" + links[i]

    # Send a GET request to the URL with the timeout
    response = requests.get(url, timeout=timeout_seconds)

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the element that contains the poem
    poem_element = soup.find('div', class_='bet-1 row pt-0 px-5 pb-4 justify-content-center')

    if poem_element:
        # Extract the text of the poem
        poems.append(poem_element.get_text())
    else:
        print("Poem not found on the website.")
```

```python
# saving each poem name with its content and poet name in a dataframe
dict1 = {'poem_name': poem_name, 'poem_content': poems}

df1 = pd.DataFrame(dict1)
df1.insert(0, 'poet_name', 'أحمد شوقي')
```

```python
# saving each poet poems in one csv file
df1.to_csv('ahmed_shawqi.csv')
```

```python
# importing all csv files for the 5 poets and combine them in one dataframe

df2 = pd.read_csv('farazdaq.csv')
df3 = pd.read_csv('albohtry.csv')
df4 = pd.read_csv('safi_al_din_al_hilli.csv')
df5 = pd.read_csv('Ibn_Arabi.csv')
df6 = pd.read_csv('ahmed_shawqi.csv')

df2 = df2.drop(df2.columns[0],axis=1)
df3 = df3.drop(df3.columns[0],axis=1)
df4 = df4.drop(df4.columns[0],axis=1)
df5 = df5.drop(df5.columns[0],axis=1)
df6 = df6.drop(df6.columns[0],axis=1)

frames = [df2, df3, df4, df5, df6]

result = pd.concat(frames).sample(frac = 1)
result.reset_index(inplace = True, drop = True)

# saving all the poems in one csv file
result.to_csv('poems.csv')
```
